```python
import numpy as np
import cv2
```

```python
x_001 = np.load('./001_fm.npy')
x_002 = np.load('./002_fm.npy')
x_003 = np.load('./003_fm.npy')
X = [x_001, x_002, x_003]
```

```python
y_001 = np.array([[1,0,0] for i in range(len(x_001))])
y_002 = np.array([[0,1,0] for i in range(len(x_002))])
y_003 = np.array([[0,0,1] for i in range(len(x_003))])
Y = [y_001, y_002, y_003]
```

```python
img_sample = x_001[180]
```

```python
import matplotlib.pyplot as plt


img_float32 = np.float32(img_sample) / 255
lab_image = cv2.cvtColor(img_float32, cv2.COLOR_BGR2RGB)


plt.imshow(lab_image)
plt.axis('off')
plt.show()
```
